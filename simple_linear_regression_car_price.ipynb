```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[6000],[8200],[9000],[14200],[16200]]).reshape(-1,1)
y = [86000,82000,78000,75000,70000]
```

```python
plt.figure()
plt.title('Car Price-KM distribution')
plt.xlabel('KM')
plt.ylabel('Price')
plt.plot(x,y,'k.')
plt.axis([3000,20000,60000,95000])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
plt.figure()
plt.title('Car Price-KM distribution')
plt.xlabel('KM')
plt.ylabel('Price')
plt.plot(x,model.predict(x),color = 'red')
plt.grid(True)
plt.show()
```

```python
test_car = np.array([[12000]])
predicted_price = model.predict(test_car)[0]
predicted_price
```

```python
test_car = np.array([[9000]])
predicted_price = model.predict(test_car)[0]
predicted_price
```

```python
y_predictions = model.predict(x)
for i,prediction in enumerate(y_predictions):
    print('Guess price %.2f, Real Price %s'% (prediction,y[i]))
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mean_absolute_error(y,y_predictions)
```

```python
mean_squared_error(y,y_predictions)
```

```python
r2_score(y,y_predictions)
```

```python
from math import sqrt

rmse = sqrt(mean_squared_error(y,y_predictions))
rmse
```
