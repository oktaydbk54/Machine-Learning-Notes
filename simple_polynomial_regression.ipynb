```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x_train = [[6000],[8200],[9000],[14200],[16200]]
y_train = [[86000],[82000],[78000],[75000],[70000]]

x_test = [[1700],[2600],[11000],[14000],[17500]]
y_test = [[94000],[94400],[73000],[83000],[75000]]
```

```python
model = LinearRegression()
model.fit(x_train,y_train)

xx = np.linspace(0,25000,150000)
yy = model.predict(xx.reshape(xx.shape[0],1))
```

```python
polynom_degree = PolynomialFeatures(degree = 2)
x_train_polynom = polynom_degree.fit_transform(x_train)
x_test_polynom = polynom_degree.transform(x_test)
```

```python
polynom_regression = LinearRegression()
polynom_regression.fit(x_train_polynom,y_train)
```

```python
xx_polynom = polynom_degree.transform(xx.reshape(xx.shape[0],1))
yy_polynom = polynom_regression.predict(xx_polynom)
xx_polynom
```

```python
polynom_regression.intercept_
```

```python
polynom_regression.coef_
```

```python
plt.figure()
plt.title('polynomial regression estimation curve')
plt.xlabel('Km')
plt.ylabel('Price')
plt.plot(xx,polynom_regression.predict(xx_polynom),c = 'r',linestyle = '-')
plt.axis([3000,22500,65000,95000])
plt.grid(True)
plt.scatter(x_train,y_train,color = 'black')
plt.show()
```

```python
plt.figure()
plt.title('polynomial regression vs linear regression')
plt.xlabel('Km')
plt.ylabel('Price')
plt.plot(xx,yy)
plt.plot(xx,polynom_regression.predict(xx_polynom),c = 'r',linestyle = '-')
plt.axis([3000,22500,65000,95000])
plt.grid(True)
plt.scatter(x_train,y_train,color = 'black')
plt.show()
```

```python
plt.figure()
plt.title('polynomial regression vs linear regression')
plt.xlabel('Km')
plt.ylabel('Price')
plt.plot(xx,yy)
plt.plot(xx,polynom_regression.predict(xx_polynom),c = 'r',linestyle = '-')
plt.axis([3000,22500,65000,95000])
plt.grid(True)
plt.scatter(x_train,y_train,color = 'black')
plt.plot(x_test,y_test,'x',color = 'black')
plt.show()
```

```python
print('simple linear regression r2: ',model.score(x_test,y_test))
print('Polynom Regression r2:',polynom_regression.score(x_test_polynom,y_test))
```
